```python
from tqdm import tqdm_notebook
from PIL import Image
import numpy as np
import os,sys
import pandas as pd
pd.options.display.max_rows = 500
pd.options.display.max_colwidth = 500
from multiprocessing import process
import cv2
import PIL
from matplotlib import pyplot as plt
import shutil
```

```python
source_folder = 'rvl-cdip/images/'
dirs = os.listdir(source_folder)
# 'rvl-cdip/images/imagese/e/j/e/eje42e00/2500126531_2500126536.tif'
len(dirs)
```

```python


def resize(img):
    if 700 <= img.size[0] <= 875:
        img = img.resize((375,500),resample=PIL.Image.LANCZOS)
    elif 500 <= img.size[0] <= 700: #reduce_height
    #     pass
#         print(img.size)
        clip_t = 120
        clip_l = 0
        box_w = img.size[0]
        box_h = img.size[1] - clip_t
#         print(clip_t,clip_l,box_w,box_h)
        img = img.resize(size=(375,500),resample=PIL.Image.LANCZOS, box = (clip_l,clip_t,box_w,box_h))
    elif 875 <= img.size[0] <= 3500: #reduce_width #too few images so not required
    #     pass
#         print(img.size)
#         clip_t = 0
#         clip_l = (img.size[0] - 875)//2
#         box_w = img.size[0] - clip_l
#         box_h = img.size[1]
#         print(clip_t,clip_l,box_w,box_h)
        img = img.resize(size=(375,500),resample=PIL.Image.LANCZOS)
    return img
```

```python
# counter = 0
img = None
size_arr = []
image_path_arr = []
save_folder='preprocessed'
for d in tqdm_notebook(sorted(dirs)):
    fol = os.path.join(source_folder,d+'/'+d[-1])
    fol_dirs = sorted(os.listdir(fol))
#     print(fol_dirs[0])
    for fd in tqdm_notebook(fol_dirs,leave=False):
        inner_fol_dir = os.path.join(fol,fd)
        inner_fol_dirs = sorted(os.listdir(inner_fol_dir))
        for ifd in inner_fol_dirs:
            inner_inner_fol_dir = os.path.join(inner_fol_dir,ifd)
            inner_inner_fol_dirs = sorted(os.listdir(inner_inner_fol_dir))
            for iifd in inner_inner_fol_dirs:
                image_folder = os.path.join(inner_inner_fol_dir,iifd)
                images = sorted(os.listdir(image_folder))
                for image in images:

                    image_path = os.path.join(image_folder,image)
    #                     print(image_path)
                    try:
                        if not os.path.exists(os.path.join(save_folder,image_folder)):
                            os.makedirs(os.path.join(save_folder,image_folder))
                        img = Image.open(fp=image_path)
#                         size_arr.append(img.size)
#                         image_path_arr.append(image_path)
                        img = resize(img)
                        img.save(os.path.join(save_folder,image_path))
                    except Exception as ex:
                        print(image_path)
                        print('Exception:',ex)
                        pass
#                         image_path_arr.append(image_path)
#                         size_arr.append((0,0))
#                             
#                     counter+=1
#                     break
#                 break
#             break
#         break
#     break

```

```python
src_folder = 'preprocessed/rvl-cdip/images/'
save_folder = 'stratified/rvl-cdip/images'
mode = ['test','val']
count = 0
for m in mode:
    paths_class = []
    df_class = pd.read_csv('rvl-cdip/labels/{0}.csv'.format(m), sep=' ', names=['path','class'])
    for g in df_class.groupby('class'):
#         print(g[0],end='\r')
        if m =='train':
            val_count = 5000
        elif m == 'val' or m == 'test':
            val_count = 1000
        for h in g[1].sort_values(by='path').values[:val_count]:
            print(h[0][:20],'-------------',count,end='\r',)
            paths_class.append(h)
            path,imgname = h[0].rsplit('/',1)
            src_path = os.path.join(src_folder,path)
            dst_path = os.path.join(save_folder,path)
#             if not os.path.exists(os.path.join(save_folder,path)):
#                 os.makedirs(os.path.join(save_folder,path))
            try:
                shutil.copytree(dst=dst_path, src=src_path)
            except Exception as ex:
                pass
    #         print(path,';',imgname,';',h[1])
            count+=1
#             break
#         break
    pd.DataFrame(paths_class).to_csv('stratified/rvl-cdip/labels/{0}.csv'.format(m),sep=' ',index=False,header=False)
#     break
count
```
